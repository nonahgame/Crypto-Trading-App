```python
import sqlite3
import pandas as pd

db_path = "/content/drive/My Drive/RBot_Data/renda_bot.db"
conn = sqlite3.connect(db_path)

step = 35
limit = 35  # Changed to match step for consistent record count
offset = 0

def display_vertical(df):
    """Display DataFrame records in vertical format"""
    if df.empty:
        print("\nNo records to display.")
        return

    print(f"\nOffset: {offset} | Showing {len(df)} records (ID range: {df['id'].iloc[0]} → {df['id'].iloc[-1]})")
    print("\n" + "="*50)

    for index, row in df.iterrows():
        print(f"Record {index + 1} (ID: {row['id']}):")
        for column, value in row.items():
            print(f"  {column}: {value}")
        print("-"*50)

while True:
    # Ensure we don't fetch beyond available records
    query = f"""
        SELECT *
        FROM trades
        ORDER BY id DESC
        LIMIT {limit} OFFSET {offset}
    """
    df = pd.read_sql_query(query, conn)

    # Display records vertically
    display_vertical(df)

    # Check if there are more records to fetch
    next_query = f"""
        SELECT COUNT(*) as count
        FROM trades
        WHERE id < {df['id'].iloc[-1] if not df.empty else 0}
    """
    next_count = pd.read_sql_query(next_query, conn)['count'].iloc[0]

    # Show navigation options based on available data
    valid_choices = []
    if offset > 0:
        valid_choices.append('[p] Previous (newer)')
    if next_count > 0:
        valid_choices.append('[n] Next (older)')
    valid_choices.append('[q] Quit')

    choice = input("\n" + " | ".join(valid_choices) + ": ").strip().lower()

    if choice == 'q':
        break
    elif choice == 'n' and next_count > 0:
        offset += step
    elif choice == 'p' and offset > 0:
        offset = max(0, offset - step)
    else:
        print("Invalid choice or no more records in that direction.")

conn.close()
```

```python
# Install required libraries (if not already available in Colab)
#!pip install pandas sqlite3 requests

import requests
import sqlite3
import pandas as pd
from datetime import datetime
#from google.colab import drive
import shutil
import os
#  download site;   https://raw.githubusercontent.com/nonahgame/td-bot/refs/heads/main/data/renda_bot.bd
# Function to download r_bot.db from GitHub  # https://raw.githubusercontent.com/nonahgame/trader-site/refs/heads/main/data/r_bot.bd
def download_renda_bot_db(url="https://raw.githubusercontent.com/nonahgame/td-bot/refs/heads/main/data/renda_bot.bd", local_path="renda_bot.db"):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(local_path, 'wb') as f:
                f.write(response.content)
            print(f"Successfully downloaded {local_path} from GitHub")
            return True
        else:
            print(f"Failed to download: {response.status_code} - {response.text}")
            return False
    except Exception as e:
        print(f"Error downloading renda_bot.db: {e}")   #.  db
        return False

# Function to save r_bot.db to Google Drive  # https://raw.githubusercontent.com/nonahgame/trader-site/main/data/r_bot.db
def save_to_google_drive(local_path="renda_bot.db", drive_folder="RBot_Data", drive_filename="renda_bot.db"):   #db
    try:
        # Mount Google Drive
        #drive.mount('/content/drive', force_remount=True)
        # Create folder in Drive if it doesn't exist
        drive_path = f"/content/drive/My Drive/{drive_folder}"
        os.makedirs(drive_path, exist_ok=True)
        # Copy file to Drive
        drive_file_path = f"{drive_path}/{drive_filename}"
        shutil.copy(local_path, drive_file_path)
        print(f"Saved {local_path} to Google Drive at {drive_file_path}")
        return drive_file_path
    except Exception as e:
        print(f"Error saving to Google Drive: {e}")
        return None

# Function to view trades from r_bot.db
def view_trades(db_path="renda_bot.db", limit=10):   #db
    try:
        conn = sqlite3.connect(db_path)
        # Query recent trades
        query = f"SELECT * FROM trades ORDER BY time DESC LIMIT {limit}"
        df = pd.read_sql_query(query, conn)
        print(f"\nRecent {len(df)} Trades:")
        display(df)

        # Summary by timeframe
        summary = pd.read_sql_query("""
            SELECT
                timeframe,
                COUNT(*) as total_trades,
                SUM(CASE WHEN action = 'buy' THEN 1 ELSE 0 END) as buy_count,
                SUM(CASE WHEN action = 'sell' THEN 1 ELSE 0 END) as sell_count,
                SUM(CASE WHEN action = 'hold' THEN 1 ELSE 0 END) as hold_count,
                SUM(CASE WHEN profit > 0 THEN 1 ELSE 0 END) as winning_trades,
                SUM(CASE WHEN profit < 0 THEN 1 ELSE 0 END) as losing_trades,
                SUM(profit) as total_profit
            FROM trades
            GROUP BY timeframe
        """, conn)
        print("\nTrade Summary by Timeframe:")
        display(summary)

        # Overall summary
        overall = pd.read_sql_query("""
            SELECT
                'All' as timeframe,
                COUNT(*) as total_trades,
                SUM(CASE WHEN action = 'buy' THEN 1 ELSE 0 END) as buy_count,
                SUM(CASE WHEN action = 'sell' THEN 1 ELSE 0 END) as sell_count,
                SUM(CASE WHEN action = 'hold' THEN 1 ELSE 0 END) as hold_count,
                SUM(CASE WHEN profit > 0 THEN 1 ELSE 0 END) as winning_trades,
                SUM(CASE WHEN profit < 0 THEN 1 ELSE 0 END) as losing_trades,
                SUM(profit) as total_profit
            FROM trades
        """, conn)
        print("\nOverall Trade Summary:")
        display(overall)

        conn.close()
        return df
    except Exception as e:
        print(f"Error retrieving database: {e}")
        return None

# Function to list tables in the database (for debugging)
def list_tables(db_path="renda_bot.db"):  #db
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        tables = cursor.fetchall()
        print("\nTables in renda_bot.db:")
        for table in tables:
            print(table[0])
        conn.close()
    except Exception as e:
        print(f"Error retrieving tables: {e}")

# Main execution
print(f"Starting at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
# Download the database
if download_renda_bot_db():  #db
    # Save to Google Drive
    drive_file_path = save_to_google_drive()
    # List tables
    list_tables()
    # View trades from local file
    view_trades()
    if drive_file_path:
        # Verify from Drive
        print("\nVerifying trades from Google Drive copy:")
        view_trades(drive_file_path)
else:
    print("Cannot proceed due to download failure. Check GitHub URL and repository.")
```
